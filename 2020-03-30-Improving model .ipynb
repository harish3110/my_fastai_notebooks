## Re-loading the model

```python
path = untar_data(URLs.FLOWERS)
df = pd.read_csv('data/df.csv', index_col=0)
```

```python
def get_x(r): return path/r['name']
def get_y(r): return r['class']

dblock = DataBlock(blocks=(ImageBlock, CategoryBlock),
                   splitter=RandomSplitter(seed=42),
                   get_x= get_x,
                   get_y= get_y, 
                   item_tfms = Resize(224))
dls = dblock.dataloaders(df) 
```

```python
learn = cnn_learner(dls, resnet34, metrics=accuracy)
```

```python
learn.load('flowers-baseline')
```

## Presizing

We have already covered how images passed into the GPU needs to have the same dimensions so that they can be collated into tensors and effectively transfereend onto to GPU and thus at the very least doing a ***Resize*** `item_fms` is mandatory. 

But if you think about it, simply resizing images is not a great idea. This might lead to producing croped off images, or images with empty zones i.e. images that aren't good. We could rotate or zoom the image to cover the space but thats just the same image interpolated and thus degraded. 

So to get over this issue presizing employs two steps:

1. 


2. 

```python
item_tfms=Resize(460),
batch_tfms=aug_transforms(size=224, min_scale=0.75)

def get_x(r): return path/r['name']
def get_y(r): return r['class']

dblock = DataBlock(blocks=(ImageBlock, CategoryBlock),
                   splitter=RandomSplitter(seed=42),
                   get_x= get_x,
                   get_y= get_y, 
                   item_tfms = item_tfms,
                   batch_tfms = batch_tfms)
dls = dblock.dataloaders(df) 
```

```python
# Sanity check
dls.show_batch(nrows=1, ncols=3)
```

```python
learn = cnn_learner(dls, resnet34, metrics=accuracy)
learn.fine_tune(3)
```

```python
learn.save('flowers-presizing')
```

We can see that Presizing improved the accuracy of the model by almost 1% (from 96% to 97%), we can see that the validation loss is also reducing and so is the training loss.
